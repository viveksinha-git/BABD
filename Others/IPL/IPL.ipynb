```python
df_matches2.to_csv('matches2.csv', index = False)
```

```python
print("Total number of matches", df_matches2.shape)
print("Total number of matches:", df_matches2.shape[0])
print("Total number of seasons:", df_matches2['season'].unique())
print("Total number of seasons:", df_matches2['season'].nunique())
```

```python
print("Maximum Player of the Match awards (Number):", df_matches2['player_of_match'].value_counts()[:10]) # Top 10 players
print()
print("Maximum Player of the Match awards (Name):", df_matches2['player_of_match'].value_counts().idxmax())
print()
print("Teams with highest number of match wins (Number):",df_matches2['winner'].value_counts()[0])
print()
print("Teams with highest number of match wins(Name):", df_matches2['winner'].value_counts().idxmax())
```

```python
# Huge Margin = 100 runs or more, 8 wickets or more

big_margin = df_matches2[(df_matches2['win_by_runs']>=100) | (df_matches2['win_by_wickets']>=8)]
big_margin.head()
print("Teams that have won matches by a huge margin:")
print(big_margin['winner'].value_counts())
```
